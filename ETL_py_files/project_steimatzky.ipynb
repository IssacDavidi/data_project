```python
# [0].find('a').get('href')
# we have 10000 hebrew books
# every page has 10 books
# which means we have 10000/10 = 1000 pages.

main_dct_steimatzky = {}


def updating_page_steimatzky(page_number):
    
    # steimatzky URL 

    url_steimatzky = 'https://www.steimatzky.co.il/%D7%A1%D7%A4%D7%A8%D7%99%D7%9D/sub-cat-all-heb-books?p={}'
    formatted_url_steimatzky = url_steimatzky.format(str(page_number))
    

    # Setting a request to the website
    request_steimatzky = requests.get(formatted_url_steimatzky)
    
    #change the format to be readable.
    soup_steimatzky = bs4.BeautifulSoup(request_steimatzky.text,'lxml')
    
    #scraping source cordinates
    target = soup_steimatzky.find_all(class_ = 'product-digital-name product-name')

    
    
       # list for the titles
    lst_title_steimatzky = [] 
    for book in range(len(target)):
        lst_title_steimatzky.append(target[book].find('a').get('title'))
    
     # list for the urls
    lst_href_steimatzky = []
    for book_href in range(len(target)):
        lst_href_steimatzky.append(target[book_href].find('a').get('href'))

    
        #dictonary zipping lists
    
        #temp dict
    temp_dct_steimatzky = dict(zip(lst_title_steimatzky,lst_href_steimatzky))
    
        #updating
    main_dct_steimatzky.update(temp_dct_steimatzky)
    
    print(f'Scraped successfully URL page number : {page_number}')
    

    

def main_steimatzky():
    
    global main_dct_steimatzky
    page_number = 0
    
    while page_number != 1000 :
        
        page_number += 1
        updating_page_steimatzky(page_number) #sub func
    print('Done web scraping!')
    
```
